```python
def plot_marginals(dataset, dataset_name):
    print(dataset_name)
    for column in dataset.columns[:-1]:
        #dataset[column] = (dataset[column] - dataset[column].mean()) / dataset[column].std()
        bins = np.linspace(dataset[column].min(), dataset[column].max(), 101)
        plt.hist(dataset[column], bins=bins)
        print(column, np.histogram(dataset[column], bins=bins))
        plt.title(column)
        plt.show()
```

```python
for dataset_name in ['communities_and_crime_v2', 'fifa19_v2', 'musk']:
    dataset = pd.read_csv(f'{datasets_dir}/{dataset_name}/dataset.csv')
    y_binary = True if dataset_name == 'musk' else False
    dataset = standardize_dataset(dataset, y_binary)
    print(dataset['y'])
    plot_marginals(dataset, dataset_name)
```
